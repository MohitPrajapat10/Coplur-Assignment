```python
# 2) DataFrames
# 	Make a Pandas DataFrame with a two-dimensional Python list 
import pandas as pd
data = [
    [1,"Mohit",200],
    [2,"Nikhil",300],
    [3,"Preeti",400]
]
df = pd.DataFrame(data, columns =  ["ID", "Name", "Marks"])
print(df)
```

```python
# Create DataFrame from Python dict
import pandas as pd
data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 21, 22],
    "City": ["Jaipur", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Create Pandas dataframe using list of lists
import pandas as pd

data = [
    ["Mohit", 20],
    ["Nikhil", 21],
    ["Preeti", 22]
]

df = pd.DataFrame(data, columns=["Name", "Age"])

print(df)
```

```python
# Create a Pandas dataframe using list of tuples 
import pandas as pd

data = [
    ("Mohit", 20),
    ("Nikhil", 21),
    ("Preeti", 22)
]

df = pd.DataFrame(data, columns=["Name", "Age"])

print(df)
```

```python
# Create a Pandas DataFrame from List of Dicts
import pandas as pd

data = [
    {"Name": "Mohit", "Age": 20},
    {"Name": "Nikhil", "Age": 21},
    {"Name": "Preeti", "Age": 22}
]

df = pd.DataFrame(data)

print(df)
```

```python
# 3)Data iteration
# 	Different ways to iterate over rows in Pandas Dataframe
import pandas as pd

data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

for index, row in df.iterrows():
    print(index, row["Name"], row["Age"])
```

```python
# Selecting rows in pandas DataFrame based on conditions
import pandas as pd

data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 18, 22]
}

df = pd.DataFrame(data)

result = df[df["Age"] > 19]

print(result)
```

```python
# Select any row from a Dataframe using iloc[]
import pandas as pd

data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

print(df.iloc[1])
```

```python
# Limited rows selection with given column
import pandas as pd

data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 21, 22],
    "City": ["Jaipur", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print(df.loc[0:1, ["Name", "City"]])
```

```python
# Drop rows from the dataframe based on certain condition applied on a column 
import pandas as pd

data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 18, 22]
}

df = pd.DataFrame(data)

df = df[df["Age"] == 18]

print(df)
```

```python
# Insert row at given position in Pandas Dataframe
import pandas as pd

data = {
    "Name": ["Mohit", "Preeti"],
    "Age": [20, 22]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Nikhil"],
    "Age": [21]
})

df1 = df.iloc[:1]
df2 = df.iloc[1:]

df = pd.concat([df1, new_row, df2]).reset_index(drop=True)

print(df)
```

```python
# Create a list from rows in Pandas dataframe
import pandas as pd

data = {
    "Name": ["Mohit", "Nikhil", "Preeti"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

row_list = df.values.tolist()

print(row_list)
```
