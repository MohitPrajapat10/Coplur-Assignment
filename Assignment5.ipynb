```python
# 1) Study the open weather API show more data in your API calling program
import requests

def weather_api(x,y):
    url = f"https://api.openweathermap.org/data/2.5/forecast?lat={x}&lon={y}&appid=2efcb5e7a008c4ef40fd85c64c26a1ac"
    response = requests.get(url)
    data = response.json()

    city = data["city"]
    city_name = city["name"]
    country = city["country"]
    city_lat = city["coord"]["lat"]
    city_lon = city["coord"]["lon"]
    return city_name, country, city_lat, city_lon

    
try:
    city_name, country, city_lat, city_lon = weather_api(24.4641, 72.7718)
    print(f"City:{city_name} \nCountry: {country} \nLatitude: {city_lat} \nLongitude: {city_lon}")
except requests.exceptions.RequestException as e:
    print(e)
```

```python
# 2) Try building today's game yourself

import random

l = ["rock", "paper", "scissor"]

while True:

    c = int(input('''
    Game Start......

    1 YES
    2 NO | Exit

    '''))

    if c == 1:

        uscore = 0
        cscore = 0

        for a in range(1, 6):

            uch = ""

            userinput = int(input('''
1 ROCK
2 PAPER
3 SCISSOR

'''))

            if userinput == 1:
                uch = "rock"

            elif userinput == 2:
                uch = "paper"

            elif userinput == 3:
                uch = "scissor"

            else:
                print("Invalid choice")
                continue

            cch = random.choice(l)

            print("Your Choice:", uch)
            print("Computer Choice:", cch)

            if uch == cch:
                print("Game Draw")
                uscore += 1
                cscore += 1

            elif (
                (uch == "rock" and cch == "scissor") or
                (uch == "paper" and cch == "rock") or
                (uch == "scissor" and cch == "paper")
            ):

                uscore += 2
                print("You Win")

            else:
                cscore += 2
                print("Computer Win")

            print("User Score:", uscore)
            print("Computer Score:", cscore)
            print("-" * 30)

        if uscore == cscore:
            print("Game Drawn")

        elif uscore > cscore:
            print("You Won the Match")

        else:
            print("Computer Won the Match")

    else:
        break
```

```python
# 3) Search for more free API's generate your and call them to fetch data. Display some data in your program.
import requests
def api():
    url = "https://api.freeapi.app/api/v1/public/randomusers/user/random"
    response = requests.get(url)
    data = response.json()

    if data["success"] and "data" in data:
        user_data = data["data"]
        username = user_data["login"]["username"]
        country = user_data["location"]["country"]
        return username,country
    else:
        raise Exception("Failed")
    
try:
    username , country = api()
    print(f"Username: {username} \nCountry: {country}")
except Exception as e:
    print(str(e))
```
