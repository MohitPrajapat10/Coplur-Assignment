```python
# 1) Create a CSV file for address book, CSV file should have column for name, address, mobile, email. Insert 2-3 dummy data entered by user.

import csv

# Create CSV file
with open("address_book.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Address", "Mobile", "Email"])

    for i in range(3):
        print(f"\nEnter details for person {i+1}:")

        name = input("Enter Name: ")
        address = input("Enter Address: ")
        mobile = input("Enter Mobile Number: ")
        email = input("Enter Email: ")

        # Writing data into CSV file
        writer.writerow([name, address, mobile, email])

print("\nData successfully stored in address_book.csv")
```

```python
# 2)  Practice DATABASE  
# 1) Create Database
# 2) Create 2-3 tables
# 3) Insert some records
# 4) Perform diffrent select operations
# 5) Update some data
# 6) Delete some data

import sqlite3

conn = sqlite3.connect("college.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    branch TEXT,
    marks INTEGER
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS teachers (
    teacher_id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT,
    duration TEXT
)
""")

cursor.execute("INSERT INTO students VALUES (1, 'Nikhil', 'CSE', 85)")
cursor.execute("INSERT INTO students VALUES (2, 'Mohit', 'IT', 78)")
cursor.execute("INSERT INTO students VALUES (3, 'Rahul', 'ECE', 90)")

cursor.execute("INSERT INTO teachers VALUES (1, 'Sharma Sir', 'Python')")
cursor.execute("INSERT INTO teachers VALUES (2, 'Verma Maam', 'DBMS')")

cursor.execute("INSERT INTO courses VALUES (1, 'Python Programming', '3 Months')")
cursor.execute("INSERT INTO courses VALUES (2, 'Database Systems', '4 Months')")

print("All Students:")
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(row)

print("\nStudents with marks greater than 80:")
cursor.execute("SELECT name, marks FROM students WHERE marks > 80")
for row in cursor.fetchall():
    print(row)

print("\nTeacher Names:")
cursor.execute("SELECT name FROM teachers")
for row in cursor.fetchall():
    print(row)

cursor.execute("""
UPDATE students
SET marks = 95
WHERE name = 'Nikhil'
""")

cursor.execute("""
DELETE FROM students
WHERE name = 'Mohit'
""")

print("\nFinal Student Table:")
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(row)

conn.commit()
conn.close()
```
